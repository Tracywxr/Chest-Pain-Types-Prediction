#### Introduction

Angina, more commonly referred to as chest pain, is the discomfort caused by the lack of oxygen-rich blood in the heart (Daly, 2006). There are a multitude of factors that have proven to increase the risk of chest pain including age, high blood pressure, smoking, physical inactivity, and more (Hermiz, 2021). There are various kinds of Angina as well,  which involves stable (typical) angina, unstable (atypical), as well different versions of symptoms that come with this heart disease. For each type of angina, the symptoms differ immensely and each symptom can represent a different level of severity (Pruthi, 2020). The purpose of our project is to predict the type of chest pain using various predictor variables. 

In order to do this, we have chosen the UCI Heart Disease data set (Aha, 2017) which examines the presence of heart disease in patients. The columns of this data set that are of interest include: age (which gives us an indication of how old the patient is), chest pain type (which is used to classify the category of angina), serum cholesterol (represents the total amount of cholesterol in the patient’s blood), maximum heart rate (measures beats per minute of the heart when it’s working at its maximum), and resting blood pressure (portrays the normal pressure at which blood flows in the patient’s circulatory system).  These variables will help us to classify which type of chest pain is present in a given patient. We will create a classification model to predict the type of angina, and then assess how well our model accurately predicts this statistic.

To start our analysis, we begin to split our dataset into the training and testing data.

```r
# split our dataset into a training dataset and a testing dataset
# Randomly take 75% of the data in the training set. 
# This will be proportional to the different number of fruit names in the dataset.
set.seed(10)

chest_pain_data_split <- initial_split(chest_pain_data, prop = 0.75, strata =chest_pain_type )  
chest_pain_data_train <- training(chest_pain_data_split)   
chest_pain_data_test <- testing(chest_pain_data_split)
```

The next step is to begin the cross validation and classification model. In this first model, we will use age and max heart rate as our predictor variables. All the predictors are standardized in our recipe and a workflow is created. We can compute the accuracy of our model then visualize it in an accuracy plot.

```r
set.seed(10) 

# cross validation
vfold <- vfold_cv(chest_pain_data_train, v=5, strata = chest_pain_type)

# model
knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) %>%
    set_engine("kknn") %>%
    set_mode("classification")

# recipe and standardization
cp_recipe <- recipe(chest_pain_type~age+max_heart_rate, data = chest_pain_data_train) %>%
    step_scale(all_predictors()) %>%
    step_center(all_predictors())

# workflow
knn_result <- workflow() %>%
    add_recipe(cp_recipe) %>%
    add_model(knn_tune) %>%
    tune_grid(resamples = vfold, grid = 50) %>%
    collect_metrics()

# accuracy
accuracies <- knn_result %>%
    filter(.metric == "accuracy")

options(repr.plot.width = 7, repr.plot.height = 6)    
# accuracy plot
cross_val_plot <- accuracies %>%
    ggplot(aes(x=neighbors, y=mean)) +
    geom_point()+
    geom_line() +
    labs(x="Neighbors", y="Accuracy Estimate") +
    ggtitle("n = 11") +
    theme(text = element_text(size = 20)) +
    scale_x_continuous(limits = c(0, 13), breaks = seq(0, 13, by = 1)) +  # adjusting the x-axis
    scale_y_continuous(limits = c(0, 1)) # adjusting the y-axis
```

Figure 3:

```r
cross_val_plot
```

Based on the accuracy plot, we will build our model using k = 11.

```r
set.seed(10)

# build model with k=11
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 11) %>%
    set_engine("kknn") %>%
    set_mode("classification")

# new workflow
fit <- workflow() %>%
    add_recipe(cp_recipe) %>%
    add_model(knn_spec) %>%
    fit(data=chest_pain_data_train)

# predict
cp_predictions <- predict(fit, chest_pain_data_test) %>%
    bind_cols(chest_pain_data_test)

# compute accuracy
metrics <- cp_predictions %>%
    metrics(truth = chest_pain_type, estimate = .pred_class)

# confusion matrix
confusion <- cp_predictions %>%
    conf_mat(truth = chest_pain_type, estimate = .pred_class)
```
